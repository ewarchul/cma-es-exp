```r
library(tidyverse)
library(magrittr)
library(cecb)
options(repr.plot.width=20, repr.plot.height=10)
```

```r
install.packages("cec2017")
```

```r
get_funcs_data = function(dim, problems) {
    problems %>%
        purrr::map_dfr(function(problem) {
            cecb::get_dfr(
                c("../data/msc-benchmarks/cec17/cma-es-csa",
                  "../data/msc-benchmarks/cec17/cma-es-expth-dp-2-pt-0.1"),
                config = list(dim = dim, probnums = problem, resp = 51)
            ) %>%
            dplyr::mutate(Func_number = problem)
        })
        
}
```

```r
DIM = 10
```

```r
unimodals =
    get_funcs_data(DIM, 1:3)
simple_unimodals = 
    get_funcs_data(DIM, 4:10)
hybrid = 
    get_funcs_data(DIM, 11:20)
comps = 
    get_funcs_data(DIM, 21:30)
    
```

# UNIMODAL FUNCTIONS

```r
unimodals %>% 
    cecb::ecdf_plot() + 
    ggplot2::facet_grid(cols = dplyr::vars(Func_number))

cecb::get_dfr(
    c("../data/msc-benchmarks/cec17/cma-es-csa",
      "../data/msc-benchmarks/cec17/cma-es-expth-dp-2-pt-0.1"),
    config = list(dim = DIM, probnums = 1:3, resp = 51)
) %>%
    cecb::ecdf_plot()
```

# SIMPLE UNIMODAL FUNCTIONS

```r
simple_unimodals %>% 
    cecb::ecdf_plot() + 
    ggplot2::facet_grid(cols = dplyr::vars(Func_number))
cecb::get_dfr(
    c("../data/msc-benchmarks/cec17/cma-es-csa",
      "../data/msc-benchmarks/cec17/cma-es-expth-dp-2-pt-0.1"),
    config = list(dim = DIM, probnums = 4:10, resp = 51)
) %>%
    cecb::ecdf_plot()
```

```r
hybrid %>% 
    cecb::ecdf_plot() + 
    ggplot2::facet_grid(cols = dplyr::vars(Func_number))
cecb::get_dfr(
    c("../data/msc-benchmarks/cec17/cma-es-csa",
      "../data/msc-benchmarks/cec17/cma-es-expth-dp-2-pt-0.1"),
    config = list(dim = DIM, probnums = 11:20, resp = 51)
) %>%
    cecb::ecdf_plot()
```
